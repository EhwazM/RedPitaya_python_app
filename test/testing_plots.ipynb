```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import bokeh.plotting as bkp
from bokeh.models import ColumnDataSource, Range1d, LinearAxis, Legend
from bokeh.layouts import column, row
from bokeh.io import curdoc

data = pd.read_csv('/home/ehwazm/Downloads/Hola_3.csv')
data.head()
```

```python
#Matplotlib Graph
fig, ax = plt.subplots(1, 1, figsize=(15, 7), layout='constrained')
ax.plot(data['x'], data['y0'], label='Canal 1')
ax.plot(data['x'], data['y1'], label='Canal 2')
ax.set_title('Señales de Canal 1 y Canal 2')
ax.set_xlabel('Tiempo (s)')
ax.set_ylabel('Amplitud (V)')
ax.legend()
plt.show() 
```

```python
# Bokeh Graph
bkp.output_notebook()
source = ColumnDataSource(data=dict(x=data['x'], y0=data['y0'], y1=data['y1']))
p = bkp.figure(title="Señales de Canal 1 y Canal 2", x_axis_label='Tiempo (s)', y_axis_label='Amplitud (V)', width=1600, height=800)
line1 = p.line('x', 'y0', source=source, line_width=2, color='blue', legend_label='Canal 1')
line2 = p.line('x', 'y1', source=source, line_width=2, color='red', legend_label='Canal 2')
p.legend.location = "top_left"
bkp.show(p)
```
